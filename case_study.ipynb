```python
data = data.dropna(subset=['address'])
```

```python
data.info()
```

```python
def remove_diacritics(text):
    if isinstance(text, str):
        return unidecode(text).upper()
    else:
        return text
```

```python
data['address'] = data['address'].apply(remove_diacritics)
data['address']
```

```python
# List of locations to filter
da_nang = ['CAM LE', 'HAI CHAU', 'LIEN CHIEU', 'NGU HANH SON', 'SON TRA', 'THANH KHE', 'HOA VANG', 'HOANG SA', 'DA NANG', 'DN']

# Regular expression pattern for matching any of the locations
pattern = '|'.join(da_nang)

# Filter addresses containing any of the specified locations
filtered_DN = data[data['address'].str.contains(pattern)]

# Display filtered addresses
filtered_DN
```

```python
def filter_location(address):
    for location in da_nang:
        if location in address:
            return location
    return None

# Display DataFrame with the new column
filtered_DN['filtered_location'] = filtered_DN['address'].apply(filter_location)
filtered_DN
```

```python
data_not_DN = data[~data['address'].str.contains(pattern)]
data_not_DN
```

```python
province_not_DN = ['Hòa Bình', 'Sơn La', 'Điện Biên', 'Lai Châu', 'Lào Cai', 'Yên Bái', 'Phú Thọ', 'Hà Giang', 'Tuyên Quang', 'Cao Bằng',
             'Bắc Kạn', 'Thái Nguyên', 'Lạng Sơn', 'Bắc Giang', 'Quảng Ninh', 'Hà Nội', 'Bắc Ninh', 'Hà Nam', 'Hải Dương', 'Hải Phòng',
             'Hưng Yên', 'Nam Định', 'Thái Bình', 'Vĩnh Phúc', 'Ninh Bình', 'Thanh Hóa', 'Nghệ An', 'Hà Tĩnh', 'Quảng Bình', 'Quảng Trị',
             'Huế', 'Quảng Nam', 'Quảng Ngãi', 'Bình Định', 'Phú Yên', 'Khánh Hòa', 'Ninh Thuận', 'Bình Thuận', 'Kon Tum', 'Gia Lai',
             'Đắk Lắk', 'Đắk Nông', 'Lâm Đồng', 'An Giang', 'Bạc Liêu', 'Bến Tre', 'Cà Mau', 'Cần Thơ', 'Đồng Tháp', 'Hậu Giang',
             'Kiên Giang', 'Long An', 'Sóc Trăng', 'Tiền Giang', 'Trà Vinh', 'Vĩnh Long']

province_not_DN = [unidecode(location).upper() for location in province_not_DN]
pattern1 = '|'.join(province_not_DN)
pattern1
```

```python
data_not_DN = data_not_DN[data_not_DN['address'].str.contains(pattern1)]
data_not_DN
```

```python
def filter_location_not_DN(address1):
    for location in province_not_DN:
        if location in address1:
            return location
    return None

# Display DataFrame with the new column
data_not_DN['filtered_location'] = data_not_DN['address'].apply(filter_location_not_DN)
data_not_DN
```
